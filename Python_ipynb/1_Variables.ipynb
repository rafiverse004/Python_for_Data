Total Salary = (Base Salary) * (1 + Bonus Rate)

```python
base_salary = 1000
bonus_rate = .10
```

```python
total_salary = (base_salary) * (1 + bonus_rate)
```

```python
total_salary
```

```python
job_id = 40
company_name = 'ABC'
job_title = 'Data analyst'
job_salary = total_salary
```
