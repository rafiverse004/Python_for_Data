What is Type Conversion

Type conversion means changing one data type into another data type.

Correct Data Conversion ✅

    int → float
    int → str
    float → int
    float → str
    str → int (only numeric string)
    str → float (only numeric string)

Incorrect Data Conversion ❌

    str ("hello") → int
    str ("python") → float
    list → int
    dict → float
