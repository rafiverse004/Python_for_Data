📌 What is a Variable

A variable is a name used to store a value in memory so it can be used later in a program.

    x = 10
    name = "Python"

🎯 Rules of variables (very important)


Variable names must start with letter or _

    my_name = "CS Daily"✅
    _my_name = "CS Daily"✅

No spaces allowed

    my name = "CS Daily"❌
    myname = "CS Daily"✅

Cannot start with number

    1myname = "CS Daily"❌
    myname1 = "CS Daily"✅

Case-sensitive

    myname = myName ❌

📌 What is a Data Type

A data type defines the kind of value a variable can store.

📌 How many Data Types in Python

Main built-in data types are:

Numeric Types

    int 10, 5
    float 10.2, 2.4
    complex (1+2i)

Text Type

    str "Hello"

Boolean Type

    bool True/False

Sequence Types

    list [1, 2, 3, "A", True]
    tuple (10, 20, 30)
    range (1,5)

Set Types

    set {1, 2, 3, 3, 2}
    frozenset([1, 2, 3, 3])

Mapping Type

    dict = {
    "name": "Alex",
    "age": 20,
    "city": "Dhaka"
    }
